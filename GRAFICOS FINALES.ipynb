```python
import numpy as np
import matplotlib.pyplot as plt

# Labels actualizadas
alternatives = [
    "Alternativa 1:\nIndividuales",
    "Alternativa 2:\nCentralizado (GMV)",
    "Alternativa 3:\nCentralizado (Inverter)",
    "Alternativa 4:\nChiller c/ Inverter"
]

# Costos (con resistencia = 0 en invierno, chiller presente en verano en 4, 5 y 6)
winter = np.array([
    273866,   # Splits
    244365,   # VRF
    336363,   # Rooftop
    386390,         # chiller
])


summer = np.array([
    233921,   # Splits
    228034,   # VRF
    339555,   # Rooftop
    397193,    # Chiller
])

# Inversin inicial
investment = np.array([
    8603298,
    5789950,
    7882524,
    8479850
])

# Clculo anual y a 10 aos
annual_cost = 4*winter + 4*summer
operation_10y = 10 * annual_cost

x = np.arange(len(alternatives))
width = 0.6

plt.figure(figsize=(12,6))
plt.grid(axis='y', zorder=0, linestyle='--', alpha=0.4)

# --- Operacin ---
plt.bar(
    x,
    operation_10y,
    width,
    label="Operacin en 10 aos",
    color="#4A90E2",
    zorder=3
)

# --- Inversin ---
plt.bar(
    x,
    investment,
    width,
    bottom=operation_10y,
    label="Inversin inicial",
    color="#4A90E2",
    alpha=0.5,
    zorder=4
)

plt.xticks(x, alternatives, rotation=45, ha="right")
plt.ylabel("Costo acumulado en 10 aos [$]")
plt.title("Costo total a 10 aos por alternativa")
plt.legend()
plt.tight_layout()
plt.show()
```
